## 💡 Contoh Data
Misalkan terdapat 4 lokasi Dragon Ball:  
(DB1, DB2, DB3, DB) </br>
Visual peta rute dan jarak antar rute dapat dilihat pada output program dibawah  

````python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.Graph()
edges = [
    ('DB1', 'DB2', 5),
    ('DB1', 'DB3', 9),
    ('DB1', 'DB4', 8),
    ('DB2', 'DB3', 7),
    ('DB2', 'DB4', 6),
    ('DB3', 'DB4', 4)
]
G.add_weighted_edges_from(edges)

pos = nx.circular_layout(G)

plt.figure(figsize=(7,7))
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color='lightblue', edgecolors='black')
nx.draw_networkx_edges(G, pos, width=2)
nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

for (n1, n2, w) in G.edges(data='weight'):
    x1, y1 = pos[n1]
    x2, y2 = pos[n2]
    xm, ym = (x1 + x2) / 2, (y1 + y2) / 2
    dx, dy = x2 - x1, y2 - y1
    norm = np.sqrt(dx**2 + dy**2)
    nx_, ny_ = -dy / norm, dx / norm
    if ny_ < 0: nx_, ny_ = -nx_, -ny_
    offset = 0.08
    plt.text(xm + nx_ * offset, ym + ny_ * offset, str(w),
             fontsize=11, fontweight='bold', color='black',
             bbox=dict(boxstyle="round,pad=0.2", fc="white", ec="none"))

plt.title("Graf Jarak Antar Dragon Ball", fontsize=14, fontweight='bold')
plt.axis('off')
plt.show()
````

<blockquote>
(2A) Jawaban :

Kemungkinan rute yang di dapatkan ada 6 kemungkinan, karena

```python
(n-1)!f
(4-1)!f
3! = 3 * 2 * 1 = 6
```

Inilah kemungkinan si Goku untuk memilih rutenya untuk mengambil rute minimum :

- Jika dimulai dari DB2 -> DB4 -> DB1 -> DB3 -> DB2 maka menghasilkan 9 + 8 + 6 + 7 = 30 rute 
- Jika dimulai dari DB2 -> DB4 -> DB3 -> DB1 -> DB2 maka menghasilkan 9 + 4 + 6 + 5 = 24 rute (Jumlah Minimum)

- Jika dimulai dari DB2 -> DB3 -> DB1 -> DB4 -> DB2 maka menghasilkan 7 + 6 + 8 + 9 = 30 rute 
- Jika dimulai dari DB2 -> DB3 -> DB4 -> DB1 -> DB2 maka menghasilkan 7 + 4 + 8 + 5 = 24 rute 

- Jika dimulai dari DB2 -> DB1 -> DB3 -> DB4 -> DB2 maka menghasilkan 5 + 6 + 4 + 9 = 24 rute 
- Jika dimulai dari DB2 -> DB1 -> DB4 -> DB3 -> DB2 maka menghasilkan 5 + 8 + 4 + 7 = 24 rute 

langkah tersebut saya lakukan agar bisa menemukan rute minimum dari pencarian Dragon ball dengan strategi Brute Force.<br>
Goku menemukan rute minimum, yaitu memulai perjalanannya dari DB1 ke DB4 dengan jumlah rute 8, selanjutnya goku pergi ke DB3 dari DB4 dengan jumlah rute 4, selanjutnya goku ke DB2 dari DB3 dengan rute 7, perjalanan terakhirnya goku pergi ke DB1 dari DB3 dengan jumlah rute 5. Jadi Goku melakukan perjalanannya dengan cepat untuk mendapatkan semua Dragon Ball dengan jumlah rute 8 + 4 + 7 + 5 = 24 rute.

Pada Brute force keleibihannya ialah mencari nilai minimum dari rutenya dengan satu persatu dengan baik. Tetapi kekurangan dari brute force yaitu pada kompleksitas waktunya O(n^2) jika data berjumlah besar dan pencarian berada di akhir dari n maka waktu yang di butuhkan untuk mencari data tersebut akan lebih lama ketimbang dengan data yang di cari berada di awal data.
