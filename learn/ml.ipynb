```python
from sklearn.neighbors import KNeighborsClassifier

# --- Langkah 1: Siapkan Data Latihan ---
# Kita punya data historis 6 siswa.
#
# Fitur (X): Jam Belajar
# Data harus dalam format 2D [[nilai]], 
# jadi [1] berarti 1 jam, [2] berarti 2 jam, dst.
X_train = [
    [1],  # Belajar 1 jam
    [2],  # Belajar 2 jam
    [3],  # Belajar 3 jam
    [5],  # Belajar 5 jam
    [6],  # Belajar 6 jam
    [7]   # Belajar 7 jam
]

# Label (y): Hasil (0 = Gagal, 1 = Lulus)
# Ini adalah "jawaban" dari data X_train
y_train = [0, 0, 0, 1, 1, 1] 
# 1 jam -> Gagal (0)
# 2 jam -> Gagal (0)
# 3 jam -> Gagal (0)
# 5 jam -> Lulus (1)
# ...dst

# --- Langkah 2: Buat dan Latih Model ---

# Kita pilih algoritma "K-Nearest Neighbors" (KNN).
# Algoritma ini memprediksi berdasarkan data "tetangga" terdekat.
# n_neighbors=3 berarti ia akan melihat 3 data terdekat.
model = KNeighborsClassifier(n_neighbors=3)

# Ini adalah inti dari "Machine Learning".
# Kita menyuruh model untuk "belajar" dari data X_train dan y_train.
model.fit(X_train, y_train)

# --- Langkah 3: Buat Prediksi ---

# Sekarang kita punya data baru: 
# Siswa A belajar selama 4 jam. Apakah dia akan lulus?
# Kita masukkan data baru (yang belum pernah dilihat model).
data_baru = [[7]] 

# Minta model untuk memprediksi hasilnya
prediksi = model.predict(data_baru)

# --- Langkah 4: Lihat Hasilnya ---

print(f"Model telah dilatih dengan data jam belajar: {X_train}")
print(f"Dan data hasil (0=Gagal, 1=Lulus): {y_train}")
print("---")
print(f"Memprediksi untuk siswa yang belajar {data_baru[0][0]} jam...")

if prediksi[0] == 1:
    print("Hasil Prediksi: Lulus (1)")
else:
    print("Hasil Prediksi: Gagal (0)")
```
